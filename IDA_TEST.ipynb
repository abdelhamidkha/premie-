```python
#objectif de modele ''en régression'' x produit en entrer (matiere en susponsion)
# y produit en sortie         r rondements 
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from keras.models import Sequential
from keras.layers import Dense, Dropout

# Load the data
df = pd.read_csv("/content/drive/My Drive/dataset/dataset.csv", delimiter= ";", decimal=".") 

# Separate the data into inputs (X and Y) and target (R)
X = df[['X', 'Y']].values
y = df['R'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Build the neural network
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=32)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate the R^2 score
r2 = r2_score(y_test, y_pred)

# Print the training and testing loss and R^2 score
print('Training Loss: ', history.history['loss'][-1])
print('Testing Loss: ', history.history['val_loss'][-1])
print('R^2 Score: ', r2)
```
